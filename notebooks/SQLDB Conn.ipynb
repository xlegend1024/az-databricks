## SQL Database connection demo

Details are here: https://docs.azuredatabricks.net/spark/latest/data-sources/sql-databases.html

```
%scala
Class.forName("com.microsoft.sqlserver.jdbc.SQLServerDriver")
```

```
%scala
val jdbcHostname = "subxdemo.database.windows.net"
val jdbcPort = 1433
val jdbcDatabase = "customerchurn"

// Create the JDBC URL without passing in the user and password parameters.
val jdbcUrl = s"jdbc:sqlserver://${jdbcHostname}:${jdbcPort};database=${jdbcDatabase}"

// Create a Properties() object to hold the parameters.
import java.util.Properties
val connectionProperties = new Properties()

connectionProperties.put("user", s"sqladmin")
connectionProperties.put("password", s"1q2w3e4R")
```

```
%scala
val driverClass = "com.microsoft.sqlserver.jdbc.SQLServerDriver"
connectionProperties.setProperty("Driver", driverClass)
```

```
%scala
val allCategory = spark.re1ad.jdbc(jdbcUrl, "[dbo].[tblCustomersChurn]", connectionProperties)
```

```
%scala
allCategory.printSchema
```

```
%scala
val customerchurn = spark.read.jdbc(jdbcUrl, "[dbo].[tblCustomersChurn]", connectionProperties)
```

```
%scala
display(customerchurn)
```
